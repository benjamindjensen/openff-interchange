```python
from openff.toolkit import ForceField, Molecule, Quantity

molecule = Molecule.from_smiles("n1cncnc1")
molecule.generate_conformers()

topology = molecule.to_topology()
topology.box_vectors = Quantity([5, 5, 5], "nanometer")

sage = ForceField("openff-2.1.1.offxml")

sage.create_interchange(topology).to_lammps(prefix="1195")
```
